```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Path to your trained model (adjust the path as needed)
model_path ="/content/drive/My Drive/PCB_Project/PCB.v2i.yolov5pytorch/best_model.pt"
# Load the YOLOv8 model
model = YOLO(model_path)

# Path to your image
image_path = '/content/drive/My Drive/PCB_Project/PCB.v2i.yolov5pytorch/test/images/VID20210601143927-96_jpg.rf.43602b457aaf6f9a0e912e3c0a0ffa2d.jpg'

# Load the image using OpenCV
image = cv2.imread(image_path)

# Run YOLO detection
results = model(image)

# Extract bounding boxes and other results
boxes = results[0].boxes  # get the boxes from the results
labels = results[0].names  # get the class names
confidence = boxes.conf  # get the confidence scores

# Iterate over each detection
for i, box in enumerate(boxes):
    x1, y1, x2, y2 = box.xyxy[0]  # get the bounding box coordinates (xyxy format)
    confidence_score = confidence[i].item()  # get confidence score
    class_id = int(box.cls[0].item())  # get the class id (assuming a single class)

    # Draw bounding box on the image
    cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)

    # Get class name
    class_name = labels[class_id]

    # Add label text (class name and confidence)
    label = f'{class_name} {confidence_score:.2f}'
    cv2.putText(image, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Convert BGR image to RGB for displaying in Jupyter/Colab
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis labels
plt.show()
```

```python
import pytesseract
import cv2
from PIL import Image

# Path to your cropped IC image
cropped_image_path = '/content/drive/My Drive/PCB_Project/cropped_ICs/IC_cropped.jpg'

# Load the image using OpenCV
image = cv2.imread(cropped_image_path)

# Convert image to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Apply binary thresholding
_, thresholded_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)

# Denoise the image
denoised_image = cv2.fastNlMeansDenoising(thresholded_image, None, 30, 7, 21)

# Resize the image for better results
height, width = image.shape[:2]
new_size = (width * 2, height * 2)
resized_image = cv2.resize(denoised_image, new_size)

# Use Tesseract with custom configuration
custom_config = r'--oem 3 --psm 6'
extracted_text = pytesseract.image_to_string(resized_image, config=custom_config)

# Print the extracted text
print("Extracted Text: ", extracted_text)
```

```python
import cv2
import easyocr
import matplotlib.pyplot as plt
import numpy as np

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'])

# Path to your image
image_path = '/content/drive/My Drive/PCB_Project/cropped_ICs/IC_cropped.jpg'

# Load the image using OpenCV (convert from BGR to RGB)
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Resize the image to a smaller size for faster processing
image_resized = cv2.resize(image_rgb, (800, 600))

# Run EasyOCR text detection
result = reader.readtext(image_resized)

# Check if any text was detected
if result:
    print("Detected Text:")
    for detection in result:
        # Print the detected text from the result list
        print(detection[1])  # detection[1] contains the detected text
else:
    print("No text detected.")

# Display the image with detected text (if any)
for detection in result:
    # Get the bounding box coordinates and the detected text
    box, text = detection[0], detection[1]

    # Draw bounding box around the detected text
    cv2.polylines(image_resized, [np.array(box)], isClosed=True, color=(0, 255, 0), thickness=2)
    cv2.putText(image_resized, text, (box[0][0], box[0][1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

# Convert the image back to BGR for displaying with OpenCV
image_resized_bgr = cv2.cvtColor(image_resized, cv2.COLOR_RGB2BGR)

# Show the image with the detected text and bounding boxes
plt.imshow(image_resized)
plt.axis('off')  # Hide axes
plt.show()

# Optionally save the image with bounding boxes
cv2.imwrite('/content/drive/My Drive/PCB_Project/annotated_image.jpg', image_resized_bgr)
```
